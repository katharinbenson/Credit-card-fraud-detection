# Data Preperation and Prediction

```python
# Determine the no.of fraud and valid cases in the dataset

fraud= data[data.Class==1]
valid= data[data.Class==0]

# Outlier
outlier_fraction = len(fraud)/len(valid)
```

```python
X=data.drop('Class',axis=1)
y=data.Class
```

```python
from sklearn.metrics import classification_report,accuracy_score
from sklearn.ensemble import IsolationForest

model= IsolationForest(max_samples=len(X),
                           contamination=outlier_fraction) #no. of ouliers we think we have
```

```python
#Fit the model
model.fit(X)
score_predict= model.decision_function(X)
y_pred= model.predict(X)

# The  prediction value for these models by default give -1 and +1 which needs to be changed to 0 and  1
y_pred[y_pred==1] =0
y_pred[y_pred==-1] =1
error_count= (y_pred!=y).sum()
print('No. of errors:', error_count)
print('Accuracy Score:',(accuracy_score(y,y_pred))*100)
print('Classification Report:\n', classification_report(y,y_pred))
```
